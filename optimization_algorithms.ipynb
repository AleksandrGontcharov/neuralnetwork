# Optimization Algorithms

```python
from neuralnetwork import Network
from neuralnetwork.datasets import nregion
from neuralnetwork.activations import absolute, relu, sigmoid
```

```python
mynetwork = Network(1)
mynetwork.add_layer(number_of_neurons = 30 , activation=absolute)
mynetwork.add_layer(number_of_neurons = 30, activation=relu)
mynetwork.add_layer(number_of_neurons = 1, activation=sigmoid)
mynetwork.summary()
```

```python
X_train, Y_train, X_val, Y_val, C, R = nregion.load_data(regions=5)
nregion.graph(X_train,Y_train)
```

```python
nregion.graph(X_train,Y_train,decision=mynetwork.decision)
```

```python
mynetwork.train_all()
```

```python
from neuralnetwork.optimizers import momentum, RMSprop, adam, nesterov_momentum

num_epochs = 50
batch_size = 64
learning_rate=0.01
num_repetitions = 5
```

```python
# Momentum
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = momentum, beta = 0.9)
```

```python
# no opimizer
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = momentum, beta = 0)
```

```python
# Nesterov Momentum
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = nesterov_momentum, beta = 0.9)
```

```python
# RMSprop
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = RMSprop, beta = 0.9)
```

```python
# Adam
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = adam, beta = 0.9)
```

```python
nregion.graph(X_train,Y_train,decision=mynetwork.decision)
```


# With regularization

```python
from neuralnetwork.regularizations import L2
# Adam
for i in range(num_repetitions):
    mynetwork.fit(X_train, Y_train, batch_size=batch_size, num_epochs=num_epochs, learning_rate=learning_rate, validation_data=(X_val, Y_val), optimizer = adam, beta = 0.9, regularization = L2)
```

```python
nregion.graph(X_train,Y_train,decision=mynetwork.decision)
```
